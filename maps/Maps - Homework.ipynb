```python
from common import iterate, logistic_wrapper
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams

sns.set_style()
%matplotlib inline
```

### (a)

```python
rcParams['figure.figsize'] = 15, 10
plt.xlabel('xlabel', fontsize=18)
plt.ylabel('ylabel', fontsize=18)
x_n_1 = iterate(
        initial_state=0.2, 
        iterations=200, 
        map_fn=logistic_wrapper(r=2)
    )

x_n_hat_1 = iterate(
    initial_state=0.200001,
    iterations=200,
    map_fn=logistic_wrapper(r=2)
)

q1_difference_1 = list(map(lambda x: abs(x[0] - x[1]), zip(x_n_1, x_n_hat_1)))
plt.grid()
plt.xlabel("n")
plt.ylabel(r"$|x_n - \hat{x}_n|$")
plt.ylim(0, 0.00001)
plt.scatter(list(range(1, 200 + 1)), q1_difference_1)
```

The trajectory of the points is sensitive to the parameter $r$.
